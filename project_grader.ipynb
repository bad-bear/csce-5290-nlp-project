```python
import pandas as pd
import os, re
import nltk
#import genism
import sklearn
import matplotlib.pyplot as mplt
import numpy as np
from collections import Counter
from nltk.corpus import stopwords as sw
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
def import_data(filename):
    new_df = pd.read_excel(filename)
    return new_df

file = 'training_set_rel3.xlsx'
print(os.path.abspath(os.getcwd()))
essay_df = import_data(file)
```

```python
def preprocess(curr_df):
    curr_df.set_index('essay_id')
    curr_df.insert(loc=28, column='unique_words', value=0)
    curr_df.insert(loc=29, column='avg_words', value=0)
    curr_df.insert(loc=30, column='filler_words', value=0)
    curr_df.insert(loc=31, column='avg_domain', value=0)
    for row_essay in curr_df['essay']:
        txt = row_essay
        txt = re.sub(r'\W+', ' ', txt)
        txt = txt.lower()
        txt = txt.replace("[^a-zA-Z]", " ")
        row_essay = txt
        #print(row_essay)
    return curr_df

essay_df = preprocess(essay_df)
```

```python
def plot_feature_tables(plot_df):
    #plot unique words
    mplt.scatter(plot_df['unique_words'], plot_df['avg_domain'], marker='o')
    mplt.title('Number of Unique Words v. Average Domain Score')
    mplt.xlabel('Unique Words')
    mplt.ylabel('Average Domain Score')
    mplt.grid()
    mplt.show()

    pass
```

```python
def feature_extraction(curr_df):
    
    num_of_uniq_words = 0
    avg_of_uniq_words = 0

    for index, row in curr_df.iterrows():
        temp_string = row['essay']
        temp_essay_list = temp_string.split(' ')
        num_of_uniq_words = len(Counter(temp_essay_list))
        row['unique_words'] = num_of_uniq_words
        avg_of_uniq_words = avg_of_uniq_words + num_of_uniq_words
    avg_of_uniq_words = avg_of_uniq_words / curr_df.shape[0]
    print('completed unique words')

    num_of_words = 0
    avg_of_words = 0
    for index, row in curr_df.iterrows():
        temp_string = row['essay']
        temp_list = temp_string.split(' ')
        num_of_words = len(temp_list)
        row['avg_words'] = num_of_words
        avg_of_words = avg_of_words + num_of_words
    avg_of_words = avg_of_words / curr_df.shape[0]
    print('completed word count')

    avg_of_filler_words = 0
    for index, row in curr_df.iterrows():
        temp_string = row['essay']
        temp_tokens = word_tokenize(temp_string)
        stopwords_x = [w for w in temp_tokens if w in sw.words('English')]
        row['filler_words'] = stopwords_x
        avg_of_filler_words = avg_of_filler_words + len(stopwords_x)
    avg_of_filler_words = avg_of_filler_words / curr_df.shape[0]
    print('completed filler words')

    for index, row in curr_df.iterrows():
        avg_domain_score = 0
        domains_added = 0

        if row['domain1_score'] is not np.nan:
            avg_domain_score += row['domain1_score']
            domains_added += 1

        if row['domain2_score'] is not np.nan:
            avg_domain_score += row['domain2_score']
            domains_added += 1

        row['avg_domain'] = avg_domain_score / domains_added
    print('completed domain score')

    plot_feature_tables(curr_df)
feature_extraction(essay_df)
```
